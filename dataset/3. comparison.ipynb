```python
# importing the needed libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Importing the data from the thre municipalities (St. Gallen, Zurich and Bern)


```python
df_sg = pd.read_excel("2. st_gallen.xlsx")
df_zh = pd.read_excel("2. zurich.xlsx")
df_be = pd.read_excel("2. bern.xlsx")
```

```python
df_sg.head()
```

#### Pandas is tends to read the index column of the xlsx file as a column. Let's drop it for all cities

```python
df_sg.drop("Unnamed: 0", axis=1, inplace=True)
df_zh.drop("Unnamed: 0", axis=1, inplace=True)
df_be.drop("Unnamed: 0", axis=1, inplace=True)
```

## 2. Comparison plots

### 2.1 Budgeted VS effective values

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5), sharex=True, sharey=True)     
fig.suptitle("Income statement per year")

# St Gallen
ax[0].plot(df_sg[df_sg["Konto"] == "Aufwand"]["Jahr"], df_sg[df_sg["Konto"] == "Aufwand"]["B J"], color="green", linewidth=2, label="Budgeted expenses")
ax[0].plot(df_sg[df_sg["Konto"] == "Ertrag"]["Jahr"], df_sg[df_sg["Konto"] == "Ertrag"]["B J"], color="turquoise", linewidth=2, label="Budgeted income")
ax[0].plot(df_sg[df_sg["Konto"] == "Aufwand"]["Jahr"], df_sg[df_sg["Konto"] == "Aufwand"]["R J"], color="red", linewidth=3, label="Actual expenses")
ax[0].plot(df_sg[df_sg["Konto"] == "Ertrag"]["Jahr"], df_sg[df_sg["Konto"] == "Ertrag"]["R J"], color="orange", linewidth=3, label="Actual income")

[ax[0].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(575, 825, 25)]

ax[0].set_title("St. Gallen")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("MLN CHF")
ax[0].legend()


# Zurich
ax[1].plot(df_zh[df_zh["Konto"] == "Aufwand"]["Jahr"], df_zh[df_zh["Konto"] == "Aufwand"]["B J"], color="green", linewidth=2, label="Budgeted expenses")
ax[1].plot(df_zh[df_zh["Konto"] == "Ertrag"]["Jahr"], df_zh[df_zh["Konto"] == "Ertrag"]["B J"], color="turquoise", linewidth=2, label="Budgeted income")
ax[1].plot(df_zh[df_zh["Konto"] == "Aufwand"]["Jahr"], df_zh[df_zh["Konto"] == "Aufwand"]["R J"], color="red", linewidth=3, label="Actual expenses")
ax[1].plot(df_zh[df_zh["Konto"] == "Ertrag"]["Jahr"], df_zh[df_zh["Konto"] == "Ertrag"]["R J"], color="orange", linewidth=3, label="Actual income")

[ax[1].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(7750, 10500, 250)]

ax[1].set_title("Zurich")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("MLN CHF")
ax[1].legend()

# Bern
ax[2].plot(df_be[df_be["Konto"] == "Aufwand"]["Jahr"], df_be[df_be["Konto"] == "Aufwand"]["B J"], color="green", linewidth=2, label="Budgeted expenses")
ax[2].plot(df_be[df_be["Konto"] == "Ertrag"]["Jahr"], df_be[df_be["Konto"] == "Ertrag"]["B J"], color="turquoise", linewidth=2, label="Budgeted income")
ax[2].plot(df_be[df_be["Konto"] == "Aufwand"]["Jahr"], df_be[df_be["Konto"] == "Aufwand"]["R J"], color="red", linewidth=3, label="Actual expenses")
ax[2].plot(df_be[df_be["Konto"] == "Ertrag"]["Jahr"], df_be[df_be["Konto"] == "Ertrag"]["R J"], color="orange", linewidth=3, label="Actual income")

ax[2].xaxis.set_ticks(range(2005, 2022+1, 2))

[ax[2].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(900, 1400, 50)]

ax[2].set_title("Bern")
ax[2].set_xlabel("Year")
ax[2].set_ylabel("MLN CHF")
ax[2].legend()

plt.show()
```

### 2.2 Discrepancies in %

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5), sharex=True, sharey=True)     
fig.suptitle("Discrepancies in %")
fig.supxlabel("year")


# St Gallen
ax[0].plot(df_sg[df_sg["Konto"] == "Aufwand"]["Jahr"], (df_sg[df_sg["Konto"] == "Aufwand"]["R J"] - df_sg[df_sg["Konto"] == "Aufwand"]["B J"]) / df_sg[df_sg["Konto"] == "Aufwand"]["R J"] * 100, color="green", linewidth=2, label="Expenses discrepancy")
ax[0].plot(df_sg[df_sg["Konto"] == "Aufwand"]["Jahr"], (df_sg[df_sg["Konto"] == "Ertrag"]["R J"] - df_sg[df_sg["Konto"] == "Ertrag"]["B J"]) / df_sg[df_sg["Konto"] == "Ertrag"]["R J"] * 100, color="turquoise", linewidth=2, label="Income discrepancy")

[ax[0].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(-2.5, 18, 2.5)]
ax[0].axhline(y = 0, color = 'orange', linestyle = '--', linewidth=2)

ax[0].xaxis.set_ticks(range(2005, 2022+1, 2))

ax[0].set_title("St. Gallen")
ax[0].set_ylabel("Percent")
ax[0].legend()


# Zurich
ax[1].plot(df_zh[df_zh["Konto"] == "Aufwand"]["Jahr"], (df_zh[df_zh["Konto"] == "Aufwand"]["R J"] - df_zh[df_zh["Konto"] == "Aufwand"]["B J"]) / df_zh[df_zh["Konto"] == "Aufwand"]["R J"] * 100, color="green", linewidth=2, label="Expenses discrepancy")
ax[1].plot(df_zh[df_zh["Konto"] == "Aufwand"]["Jahr"], (df_zh[df_zh["Konto"] == "Ertrag"]["R J"] - df_zh[df_zh["Konto"] == "Ertrag"]["B J"]) / df_zh[df_zh["Konto"] == "Ertrag"]["R J"] * 100, color="turquoise", linewidth=2, label="Income discrepancy")

[ax[1].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(-2, 10, 2)]
ax[1].axhline(y = 0, color = 'orange', linestyle = '--', linewidth=2)

ax[1].xaxis.set_ticks(range(2005, 2022+1, 2))

ax[1].set_title("Zurich")

# Bern
ax[2].plot(df_be[df_be["Konto"] == "Aufwand"]["Jahr"], (df_be[df_be["Konto"] == "Aufwand"]["R J"] - df_be[df_be["Konto"] == "Aufwand"]["B J"]) / df_be[df_be["Konto"] == "Aufwand"]["R J"] * 100, color="green", linewidth=2, label="Expenses discrepancy")
ax[2].plot(df_be[df_be["Konto"] == "Aufwand"]["Jahr"], (df_be[df_be["Konto"] == "Ertrag"]["R J"] - df_be[df_be["Konto"] == "Ertrag"]["B J"]) / df_be[df_be["Konto"] == "Ertrag"]["R J"] * 100, color="turquoise", linewidth=2, label="Income discrepancy")

[ax[2].axhline(y=i, color='darkgrey', linestyle='--') for i in np.arange(-2, 10, 2)]
ax[2].axhline(y = 0, color = 'orange', linestyle = '--', linewidth=2)


ax[2].xaxis.set_ticks(range(2005, 2022+1, 2))

ax[2].set_title("Bern")

plt.show()
```
